```python
from ultralytics import YOLO

# Load the YOLOv8 model
model = YOLO('yolov8n.pt')

# Train the model using the corrected YAML file
model.train(data='/kaggle/input/football-play-detect/data.yaml', epochs=30, imgsz=640)
```

```python
results = model.predict(source='/kaggle/input/playing/R.jpg', save=True)
```

```python
import os

output_dir = '/kaggle/working/runs/detect/train22'
os.listdir(output_dir)
```

```python
from IPython.display import Image, display

for file in os.listdir(output_dir):
    if file.endswith('.jpg'):  # Filter for image files
        display(Image(filename=os.path.join(output_dir, file)))
```

```python

# Function to display images with predictions
def display_images_with_predictions(output_dir):
    for file in os.listdir(output_dir):
        if file.endswith('.jpg'):  # Filter for image files
            img_path = os.path.join(output_dir, file)
            img = cv2.imread(img_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for correct color display
            
            plt.figure(figsize=(10, 10))
            plt.imshow(img)
            plt.axis('off')  # Hide axes
            plt.title('YOLOv8 Prediction')
            plt.show()

            # Display bounding box details
            predictions = results[0].boxes  # Get predictions
            for box in predictions:
                x1, y1, x2, y2, conf, cls = box.xyxy[0].tolist() + [box.conf[0].item(), int(box.cls[0].item())]
                label = f'Class: {model.names[cls]}, Conf: {conf:.2f}'
                # Draw bounding box on the image
                cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 2)
                cv2.putText(img, label, (int(x1), int(y1 - 10)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

            # Display the annotated image
            plt.figure(figsize=(10, 10))
            plt.imshow(img)
            plt.axis('off')  # Hide axes
            plt.title('YOLOv8 Predictions with Bounding Boxes')
            plt.show()

# Call the function to display images with predictions
display_images_with_predictions(output_dir)

# Function to plot histograms of predicted classes
def plot_class_histogram(results):
    class_ids = [int(box.cls[0].item()) for box in results[0].boxes]
    class_names = [model.names[c] for c in class_ids]
    
    plt.figure(figsize=(12, 6))
    plt.hist(class_names, bins=len(model.names), color='blue', alpha=0.7)
    plt.title('Class Distribution of Predictions')
    plt.xlabel('Classes')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.show()

# Call the function to plot the histogram
plot_class_histogram(results)
```
